# 5 BSs

### Process Links Files

```python
import json
import math

with open('./T2_5_links.json', 'r') as node_file:
    json_input = node_file.read()

link_data = json.loads(json_input)
eepran_link_data = {'links': []}

for link in link_data['links']:
    eepran_link_data['links'].append({
        'Node1': link['fromNode'], 
        'Node2': link['toNode'], 
        'Delay': link['delay'] * 0.005,
        'PortCapacity': 100,
        'NumLinks': math.ceil(link['capacity'] / 100),
        'PluggableTransceiverPower': 4.5,
        'SwitchPortPower': 14 
    })

with open('./EEPRAN_T2_5_links.json', 'w') as link_file:
    link_file.write(json.dumps(eepran_link_data, indent=4))
```

### Aggregate nodes by number of links from the core

```python
import json

with open('./T2_5_links.json', 'r') as node_file:
    json_input = node_file.read()

link_data = json.loads(json_input)
node_levels = {0: [0], 1: []}
current_node_level = 1
upper_node_level = node_levels[current_node_level-1].copy()

for link in link_data['links']:

    # if link['fromNode'] in node_levels[current_node_level]:
    #     continue

    if link['fromNode'] not in upper_node_level:
        # -- We changed level --
        current_node_level += 1
        upper_node_level += node_levels[current_node_level-1]
        node_levels[current_node_level] = []

    # -- We are in current level --
    # Adding destination node to current level list
    if link['toNode'] not in node_levels[current_node_level]:
        node_levels[current_node_level].append(link['toNode'])

node_levels
```

### Process Nodes File

```python
import json
import random

hardwares = [1, 2]

json_input = ''
with open('./T2_5_CRs.json', 'r') as node_file:
    json_input = node_file.read()

        
node_data = json.loads(json_input)
eepran_node_data = {'nodes': []}

for node in node_data['nodes']:
    eepran_node_data['nodes'].append({
        'Number': node['nodeNumber'],
        'Hardwares': [hardwares[random.randint(0, len(hardwares)-1)]
                      for i in range(2) 
                      if node['nodeNumber'] in node_levels[1]],
        'StaticPercentage': random.randint(20, 25) / 100.0 
                            if node['nodeNumber'] in node_levels[1] else 0.0,
        'BaseStations': [1]
                        if node['nodeNumber'] != 0 else []       
    })

with open('./EEPRAN_T2_5_nodes.json', 'w') as node_file:
    node_file.write(json.dumps(eepran_node_data, indent=4))
```
